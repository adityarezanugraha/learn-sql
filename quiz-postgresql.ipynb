#### 1. Buatlah query untuk menampilkan seluruh data karyawan dari <b>table employees</b> yang <b>diurutkan berdasarkan email paling terakhir</b>. 

```python
%sql $postgre_con SELECT * FROM employees ORDER BY email DESC;
```

#### 2. Buatlah query untuk menampilkan data karyawan yang <b>gajinya lebih besar 3200.00 sampai dengan 12000.00.</b>

```python
%sql $postgre_con SELECT employee_id, last_name, email, salary FROM employees WHERE salary BETWEEN 3200 AND 12000;
```

#### 3. Buatlah query untuk menampilkan data karyawan yang <b>memiliki huruf A diawal nama depannya</b>.

```python
%sql $postgre_con SELECT employee_id, first_name, email, salary FROM employees WHERE first_name LIKE 'A%';
```

#### 4. Buatlah query untuk menampilkan data karyawan yang memiliki <b>kode karyawan diantaranya 103, 115, 196, 187, 102 dan 100</b>

```python
%sql $postgre_con SELECT employee_id, first_name FROM employees WHERE employee_id IN (103, 115, 196, 187, 102, 100);
```

#### 5. Buatlah query untuk menampilkan data karyawan yang <b>nama belakangnya memiliki huruf kedua u</b>

```python
%sql $postgre_con SELECT employee_id, last_name FROM employees WHERE last_name LIKE '_u%';
```

#### 6. Buatlah query untuk menampilkan <b>kode department apa saja yang ada di tabel employees secara unique</b>

```python
%sql $postgre_con SELECT DISTINCT(department_id) FROM employees;
```

#### 7. Buatlah query untuk menampilkan nama lengkap karyawan, kode jabatan, gaji setahun dari table employees yang <b>kode manager sama dengan 100</b>.

```python
%sql $postgre_con SELECT first_name || ' ' || last_name AS nama_lengkap, job_id AS kode_jabatan, salary * 12 AS gaji_setahun FROM employees WHERE manager_id = 100;
```

#### 8. Buatlah query untuk menampilkan nama belakang, gaji perbulan, kode jabatan dari table employees yang <b>tidak memiliki komisi</b>

```python
%sql $postgre_con SELECT last_name AS nama_belakang, salary AS gaji_perbulan, job_id AS kode_jabatan FROM employees WHERE commission_pct IS NULL;
```

#### 9. Buatlah query untuk menampilkan data karyawan yang <b>bukan dari jabatan IT_PROG dan SH_CLERK</b>.

```python
%sql $postgre_con SELECT employee_id, first_name, email, salary FROM employees WHERE NOT(job_id = 'IT_PROG') AND NOT(job_id = 'SH_CLERK');
```
